```python
import random
def infect(word):

    nukta_map              =    {'क़':'क',
                                'ख':'ख़',
                                'ग':'ग़',
                                'ज़':'ज',
                                'ड़':'ड',
                                'ढ़':'ढ',
                                'फ़':'फ',
                                'य़':'य',
                                'ळ':'ऴ',
                                'ऩ':'न'}
    diacritic_map           =   {'ो':'ाे',
                                'ौ':'ाै',
                                'आ':'अा',
                                'ओ':'अो',
                                'औ':'अौ',
                                'ऐ':'एे'}


    vowel_diacritics       =  ['ऺ', 'ऻ','ा', 'ि', 'ी', 'ु', 'ू', 'ृ', 'ॄ', 'ॅ', 'ॆ', 'े', 'ै', 'ॉ', 'ॊ', 'ो', 'ौ','ॎ', 'ॏ','ॕ','ॖ', 'ॗ']

    
    vowels                 =   ['ऄ', 'अ', 'आ', 'इ', 'ई', 'उ', 'ऊ', 'ऋ', 'ऌ', 'ऍ', 'ऎ', 'ए', 'ऐ', 'ऑ', 
                                'ऒ', 'ओ', 'औ','ॠ', 'ॡ', 'ॢ', 'ॣ','ॲ','ॳ', 'ॴ', 'ॵ','ॶ', 'ॷ']

    consonants             =  ['क', 'ख', 'ग', 'घ', 'ङ', 'च', 'छ', 'ज', 'झ', 'ञ', 'ट', 'ठ', 'ड', 'ढ', 'ण', 'त', 'थ', 'द', 'ध', 'न', 'ऩ', 'प', 'फ', 'ब', 
                                'भ', 'म', 'य', 'र', 'ऱ', 'ल', 'ळ', 'ऴ', 'व', 'श', 'ष', 'स', 'ह','क़', 'ख़', 'ग़', 'ज़', 'ड़', 'ढ़', 'फ़', 'य़','ॸ', 'ॹ', 'ॺ']


    connector              =   '्'
    nukta                  =   '़'
    original="".join([c for c in word])
    try:    
        part = [i for i in word]
        new_word = []
        # add connector
        if random.randint(0, 100) > 70:
            index_connector = random.randint(1, len(word)-1)
            word = word[:index_connector]+ word[index_connector:]
            index_connector = random.randint(1, len(word)-1)
            word = word[:index_connector] + connector + word[index_connector:]
        
        # adding vowel with vowel diacs
        for i in word:
            if i in vowels:
                if random.randint(0, 100) > 50:
                    word = word.replace(i, i+vowel_diacritics[random.randint(0, len(vowel_diacritics)-1)])
        # break nukta
        for i in part:
            for k in list(nukta_map):
                if i == k:
                    if random.randint(0, 100) > 50:
                        word = word.replace(k, nukta_map[k])
        # break diacritics
        for i in range(len(part)):
            for k in list(diacritic_map):
                if part[i] == k:
                    if random.randint(0, 100) > 50:
                        d_part = [_ for _ in diacritic_map[k]]
                        word = ("".join(part[:i]) + d_part[0] + d_part[1] + "".join(part[i+1:]))


        return word
    except Exception as e:
        return original
```

```python
csvs=[csv for csv in glob(os.path.join("/kaggle/input/gpun-iitp-hindi-sentiment-data/iitp-product-reviews/","*/*.csv"))]
dfs=[pd.read_csv(csv,names=['sentiment','sentence']) for csv in csvs]
df=pd.concat(dfs,ignore_index=True)
# collect unique words
unique_words=[]
total_words=0
for idx in tqdm(range(len(df))):
    sen=df.iloc[idx,-1]
    words=sen.split()
    total_words+=len(words)
    for word in words:
        if word not in unique_words: unique_words.append(word)
print("Total Words:",total_words)
print("Unique Words:",len(unique_words))
```

```python
# wom
wom=[bnorm(word)["normalized"] for word in tqdm(unique_words)]
# wim
wim=[inorm.normalize(word) for word in tqdm(unique_words)]
# un1
un1=[infect(word) for word in tqdm(unique_words)]
# un2
un2=[infect(word) for word in tqdm(un1)]
# un5
un3=[infect(word) for word in tqdm(un2)]
un4=[infect(word) for word in tqdm(un3)]
un5=[infect(word) for word in tqdm(un4)]

word_map=pd.DataFrame({"word":unique_words,
                       "ours":wom,
                       "indic":wim,
                       "un-norm(1st-deg)":un1,
                       "un-norm(2nd-deg)":un2,
                       "un-norm(5th-deg)":un5})
word_map
```

```python
# create map for dataset
def create_input_map(df,colnum,norm_method):
    norm_map={}
    for idx in tqdm(range(len(df))):
        word=df.iloc[idx,0]
        out=df.iloc[idx,colnum]
        # if something is none
        if out is None:
            norm_map[word]=word
            continue
        
        if norm_method==None:
            norm_map[word]=out
        elif norm_method=="om":
            norm=bnorm(out)["normalized"]
            if norm is None:
                norm_map[word]=word
            else:
                norm_map[word]=norm
        elif norm_method=="im":
            norm=inorm.normalize(out)
            norm_map[word]=norm
    return norm_map

# clean sentence
def clean_sentence(sent,norm_map):
    sents=[]
    words=sent.split()
    for w in words:
        try:
            n=norm_map[w]
            sents.append(n)
        except Exception as e:
            print(e,w)
            sents.append(w)
    return " ".join(sents)

# create dataset
def create_df(data,norm_map,colname):
    df=data.copy()
    df[colname]=df[colname].progress_apply(lambda x:clean_sentence(x,norm_map))
    return df
    
```

```python
import numpy as np
from sklearn.metrics import f1_score

def get_scores(model,data,iden):
    result, model_outputs, wrong_predictions = model.eval_model(data)
    bads = {}
    for pred in wrong_predictions:
        if pred.label in bads:
            bads[pred.label] += 1
        else:
            bads[pred.label] = 1
    score=round(1-(len(wrong_predictions)/len(data)),4)
    preds=np.argmax(model_outputs,axis=1)
    labels=np.array(data.labels.tolist())
    F1=f1_score(labels,preds, average='weighted')
    return {"score":score,"F1":F1,"mcc":result["mcc"]}
```
